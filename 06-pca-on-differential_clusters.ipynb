```python
import os
import pandas as pd
import r_pca
```

```python
to_save=True
```

# Get Data

```python
retina_clusters_expression_subset = pd.read_csv('data/retina/differential_clusters_expression.csv',
                                                index_col=0)
retina_clusters_expression_subset[['NRXN2', 'ATP1B1', 'PAX6', 'SLC32A1', 'SLC6A1', 'ELAVL3']]
```

# PCA

```python
# Warning: this takes a long time...
smusher = r_pca.R_pca(retina_clusters_expression_subset.to_numpy())
%time lowrank, sparse = smusher.fit(max_iter=200)  # maybe change this to 105 to match Olga's notebook?
lowrank
```

# Lowrank

```python
lowrank_df = pd.DataFrame(lowrank,
                          index=retina_clusters_expression_subset.index,
                          columns=retina_clusters_expression_subset.columns)
lowrank_df.index.name = 'cell'
print(lowrank_df.shape)
lowrank_df.head()
```

```python
if to_save:
    os.makedirs("data/pca", exist_ok=True)
    lowrank_df.to_csv('data/pca/differential_clusters_pca_lowrank.csv', index=True)
```
