```python
import pandas as pd
import numpy as np
import r_pca
```

```python
to_save=False
```

# Get Data

```python
cell_metadata = pd.read_csv('data/metadata/cell_metadata.csv', index_col=0, squeeze=True)
gene_metadata_amacrine = pd.read_csv('data/amacrine/amacrine_gene_metadata.csv', index_col=0)
table1_diff_genes_cells = pd.read_csv('data/differential_clusters/differential_clusters_expression.csv',
                                     index_col=0)
```

```python
table1_diff_genes_cells[['NRXN2', 'ATP1B1', 'PAX6', 'SLC32A1', 'SLC6A1', 'ELAVL3']]
```

# PCA

```python
# Warning: this takes a long time...
smusher = r_pca.R_pca(table1_diff_genes_cells.to_numpy())
%time lowrank, sparse = smusher.fit(max_iter=200)  # maybe change this to 105 to match Olga's notebook?
lowrank
```

```python
lowrank_df = pd.DataFrame(lowrank,
                          index=table1_diff_genes_cells.index,
                          columns=table1_diff_genes_cells.columns)
print(lowrank_df.shape)
lowrank_df.head()
```

```python
if to_save:
    lowrank_df.to_csv('data/differential_clusters_lowrank.csv', index=True)
```

# Sparse DF

```python
sparse_df = pd.DataFrame(sparse, index=table1_diff_genes_cells.index, columns=table1_diff_genes_cells.columns)
print(sparse_df.shape)
sparse_df.head()
```

```python
if to_save:
    sparse_df.to_csv('data/differential_clusters/pca/differential_clusters_lowrank_sparse.csv')
```

# Tidy

```python
def tidify_and_log(data):
    tidy = data.unstack().reset_index()
    tidy = tidy.rename(columns={'level_1': 'barcode', 'level_0': 'gene_symbol', 0: 'expression'})
    tidy['expression_log'] = np.log10(tidy['expression'] + 1)
    return tidy
```

```python
lowrank_tidy = tidify_and_log(lowrank_df)
print(lowrank_tidy.shape)
lowrank_tidy.head()
```

```python
if to_save:
    lowrank_tidy.to_csv('data/differential_clusters/pca/differential_clusters_lowrank_tidy.csv')
```

# Metadata

```python
lowrank_tidy_metadata = lowrank_tidy.join(cell_metadata, on='barcode')
print(lowrank_tidy_metadata.shape)
lowrank_tidy_metadata.head()
```

```python
if to_save:
    lowrank_tidy_metadata.to_csv('data/differential_clusters/pca/differential_clusters_lowrank_tidy_metadata.csv')
```

# Amacrine

```python
rows = lowrank_tidy_metadata['gene_symbol'].isin(gene_metadata_amacrine.index)
lowrank_tidy_amacrine = lowrank_tidy_metadata.loc[rows]
lowrank_tidy_amacrine = lowrank_tidy_amacrine.query('celltype == "Amacrine cells"')
print(lowrank_tidy_amacrine.shape)
lowrank_tidy_amacrine.head()
```

```python
if to_save:
    lowrank_tidy_amacrine.to_csv('data/differential_clusters/pca/differential_clusters_lowrank_tidy_metadata_amacrine.csv')
```
